```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
Q_in = 30000
c = 3 * 10**8
wavelen = 1550e-9
freq = c / wavelen
radius = 50e-6
roundtrip_len = 200e-6 * 2 + 2 * np.pi * radius
# roundtrip_len = 2 * np.pi * radius
v_g = 2.7
roundtrip_time = roundtrip_len / (c / v_g)
loss = 2 * np.pi * freq * roundtrip_time / Q_in
print(loss)
```

```python
kappa = np.linspace(0,0.99, 1000)
alpha = 1-loss
amp = kappa / (1 - alpha * np.sqrt(1 - np.square(kappa)))
energy = np.square(amp)
print('alpha square', alpha**2)

plt.plot(kappa**2, energy)
```
